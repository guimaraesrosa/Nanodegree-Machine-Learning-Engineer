## Análise e Exploração dos dados

```python
#Bibliotecas necessárias para este projeto
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#mostrar gráficos em notebook
%matplotlib inline
```

```python
pd.set_option('display.max_rows', 806)

#carregar arquivo csv para dataframe
df_trombose = pd.read_csv("TSUM_B.CSV")
df_trombose
```

```python
df_trombose.shape
```

```python
df_trombose.index
```

```python
#contagem de dados não nulos
df_trombose.count()
```

```python
pd.set_option('display.max_rows', 806)

#Removendo colunas indesejadas
df_trombose = df_trombose.drop(["ID","Examination Date","Symptoms"], axis=1)
df_trombose
```

```python
#Verifica novamente a quantidade de colunas e linhas
df_trombose.shape
```

```python
#tipos de dados do dataframe
df_trombose.dtypes
```

```python
#Quantidade de valores não nulos da nova base de dados
df_trombose.notnull().sum()
```

```python
#Visualização das classes
fig, ax = plt.subplots()
df_trombose['Thrombosis'].value_counts().plot(ax=ax, kind='bar', title ="", figsize=(15, 10), legend=True, fontsize=12)

ax.set_xlabel("Classes", fontsize=12)
ax.set_ylabel("Quantidade", fontsize=12)
plt.show()
```

```python
df_trombose['Thrombosis'].value_counts()
```

```python
pd.set_option('display.max_rows', 806)

df_trombose['Diagnosis'] = df_trombose['Diagnosis'].str.replace('+',',')
df_trombose
```

```python
pd.set_option('display.max_rows', 806)

df_trombose['Diagnosis'] = df_trombose['Diagnosis'].str.replace('susp',' ')
df_trombose
```

```python
#Salvando dataframe como csv

df_trombose.to_csv('exploration_data_trombose.csv', index=False)
```
